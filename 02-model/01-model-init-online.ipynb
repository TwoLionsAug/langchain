# 调用线上模型

```python
import os
from dotenv import load_dotenv
from langchain_community.chat_models import ChatZhipuAI
from langchain_deepseek import ChatDeepSeek

# 引入环境变量
load_dotenv(override=True)
DEEPSEEK_DEFAULT_MODEL = os.getenv("DEEPSEEK_DEFAULT_MODEL")
DEEPSEEK_API_KEY = os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL = os.getenv("DEEPSEEK_BASE_URL")


llm_deepseek = ChatDeepSeek(
    model = DEEPSEEK_DEFAULT_MODEL,
    api_key = DEEPSEEK_API_KEY,
    base_url = DEEPSEEK_BASE_URL,
)

response = llm_deepseek.invoke("你是什么人？")

print(response)
```

## 智谱api调用

```python
import os
from dotenv import load_dotenv
from langchain_community.chat_models import ChatZhipuAI
# 引入环境变量
load_dotenv(override=True)
ZHIPUAI_DEFAULT_MODEL = os.getenv("ZHIPUAI_DEFAULT_MODEL")
ZHIPUAI_API_KEY = os.getenv("ZHIPUAI_API_KEY")
ZHIPUAI_BASE_URL = os.getenv("ZHIPUAI_BASE_URL")


llm_zhipu= ChatZhipuAI(
    model = "glm-5.1",
    api_key = ZHIPUAI_API_KEY,
    api_base = ZHIPUAI_BASE_URL,
)

response = llm_zhipu.invoke("你是谁")

print(response)
```
